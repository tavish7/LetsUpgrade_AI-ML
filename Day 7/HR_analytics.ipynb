### There are many categorical values so we'll use LabelEncoder to make it numerical

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#encodes'Yes'=1, 'No'=0
df3['Attrition'] = le.fit_transform(df3['Attrition'])

#encodes 'Travel_Rarely'=2, 'Travel_Frequently'=1, 'Non-Tavel'=0
df3['BusinessTravel'] = le.fit_transform(df3['BusinessTravel'])

#encodes 'Sales'=2,'Research & Development'=1,'Human Resources'=0 for Department column
df3['Department'] = le.fit_transform(df3['Department'])

#encodes 'Technical Degree'=5,'Other'=4,'Medical'=3,'Marketing'=2,'Life Sciences'=1,'Human Resources'=0 for EducationField column
df3['EducationField'] = le.fit_transform(df3['EducationField'])

#encodes 'Male'=1,'Female'=0 for Gender column
df3['Gender'] = le.fit_transform(df3['Gender'])

#encodes 'Sales Representative'=8,'Sales Executive'=7,'Research Scientist'=6,'Research Director'=5,'Manufacturing Director'=4,'Manager'=3,'Laboratory Technician'=2,'Human Resources'=1,'Healthcare Representative'=0 in JobRole column
df3['JobRole'] = le.fit_transform(df3['JobRole'])

#encodes 'Single'=2,'Married'=1,'Divorced'=0 in MaritalStatus column
df3['MaritalStatus'] = le.fit_transform(df3['MaritalStatus'])
```

```python
df3
```

```python
df3.corr()
```

```python
df3.dropna()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df3.corr(),annot=True)
```
