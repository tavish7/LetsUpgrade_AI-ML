```python
# drop irrelevant data
df1 = df.drop(['ID','ZIP Code'], axis=1)
df1.head()
```

```python
x = df1.drop(['Personal Loan'],axis=1)
x.head()
```

```python
y = df.iloc[:,9:10]
y.head()
```

# Creating Logistic Regression Model

```python
import statsmodels.api as sm
```

```python
x1=sm.add_constant(x)
logistic=sm.Logit(y,x1)
result=logistic.fit()
result.summary()
```

__Inference__

Here all the independent variables have ___p-value < 0.05,___ which means all the variables play significant role in predicting whether customer accept the personal loan or not.
Out of these ___Age, Experience, Mortgage___ play most significant role.
