```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('general_data.csv')
df.head()
```

```python
df['Attrition'].replace(to_replace=('No','Yes'),value=(0,1), inplace=True)
df['Gender'].replace(to_replace=('Male','Female'),value=(0,1), inplace=True)
df.head()
```

```python
# 1. Correlation between Attrition and Age
# Ho : There is no significance difference between Attrition and Age

stats,p = pearsonr(df.Attrition,df.Age)
print("Correlation : ",stats," P-value : ",p)
if p>=0.05:
    print("Accept Null Hypothesis")
elif p<0.05:
    print("Reject Null Hypothesis")
if stats>0:
    print("Positive Correlation")
elif stats<0:
    print("Negative Correlation")
```
