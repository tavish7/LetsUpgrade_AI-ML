```python
df1 = df.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df1["Sex"] = le.fit_transform(df1["Sex"])
df1["Embarked"] = le.fit_transform(df1["Embarked"])
```

```python
df1.head()
```

# Naive Bayes

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import * 
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
def nb(col):    
    clf = BernoulliNB()
    for i in col:
        x = df1.drop([i], axis=1)
        y = df1[i]
        llist = df1.columns.to_list()
        llist.remove(i)
        print('Dependent Variable:', i)
        print('Independent Variables:', *llist)
    
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.3, random_state=0)
        model = clf.fit(x_train,y_train)
        pred = model.predict(x_test)
    
        print(f"\nAcuuracy for {i} is: ", accuracy_score(y_test, pred, normalize=True))
        print(f"Confusion matrix for {i} is \n", confusion_matrix(y_test, pred))
        print("===================================================")
```

```python
col = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch','Embarked']        
nb(col)
```
