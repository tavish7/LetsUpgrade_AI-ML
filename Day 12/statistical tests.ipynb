```python
import pandas as pd
from scipy.stats import mannwhitneyu
from scipy.stats import chi2_contingency

df = pd.read_csv('general_data.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df=df.dropna()
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.isna().sum()
```

```python
att_y=df[df['Attrition']=='Yes']
att_n=df[df['Attrition']=='No']
```

```python
att_y
```

```python
att_n
```

# Statistical Analysis

## 1. Attrition VS Age

__H0:__ There is no significant difference in the employee's age between the employee who left and the employee who stayed. 

__H1:__ There is significant difference in the employee's age between the employee who left and the employee who stayed.



```python
stats1,p1=mannwhitneyu(att_y.Age,att_n.Age)
print("Critical value:",stats1)
print("p value:",p1)
if p1<0.05:
    print("Reject H0")
elif p1>0.05:
    print("Accept H0")
```

Since H0 is rejected which implies, _There is significant difference in the employee's age between the employee who left and the employee who stayed._

***

## 2. Attrition VS BusinessTravel

__H0:__ There is no significant difference in the employee's Attrition w.r.t BussinessTravel.

__H1:__ There is significant difference in the employee's Attrition w.r.t BussinessTravel.

```python
chitable1=pd.crosstab(df.Attrition,df.BusinessTravel)
chitable1
```

```python
stats,p2,dof,expeted=chi2_contingency(chitable1)
print("Critical value:",stats)
print("p value:",p2)
if p2<0.05:
    print("Reject H0")
elif p2>0.05:
    print("Accept H0")
```

Since H0 is rejected which implies, _There is significant difference in the employee's Attrition w.r.t BussinessTravel._
